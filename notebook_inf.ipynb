# Model Inference

1. Import Libarary

```python
# Import Libraries

import pickle
import pandas as pd
```

2. Load

Load file dengan pickle.

```python
# Load

with open('model_best.pkl', 'rb') as file:
    best_pipe = pickle.load(file)
```

3. Data Inference

Terdapat 2 user baru yang akan diprediksi apakah user `will_buy_on_return_visit`

```python
# Data inference

data_inf_input = {
    'timestamp': '2025-06-18 00:06:00',
    'machine_id': 18,
    'temperature': 89,
    'vibration': 85,
    'humidity': 70,
    'pressure': 1.5,
    'energy_consumption': 3.6,
    'machine_status': 1,
    'anomaly_flag': 1,
    'predicted_remaining_life': 80,
    'failure_type': 'Overheating',
    'downtime_risk': 0.7,
}

# Data frame

data_inference = pd.DataFrame([data_inf_input])
print(data_inference)
```

```python
# Prediction (0/1)

pred = best_pipe.predict(data_inference)

# Probability prediction for 1

proba = best_pipe.predict_proba(data_inference)[:, 1]

# Result

result = data_inference.copy()
result['maintenance_required'] = pred
result['predicted_probability'] = proba

print(result)
```

The new data or inference of equipment is 1 or need maintenance with probability around 73.6%, it needs predictive maintenance before breakdown occur.
